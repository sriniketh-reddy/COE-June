```python
import pandas as pd
```

```python
df=pd.read_pickle('/content/2_CarPricesData.pkl')
```

```python
df.head()
```

```python
X=df.iloc[:,:-1]
y=df['Price']
```

```python
X.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
df[['Age','KM','Weight','HP','CC']]=sc.fit_transform(df[['Age','KM','Weight','HP','CC']])
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42)
```

```python
import tensorflow
from keras import Sequential
from keras.layers import Dense
```

```python
model=Sequential()
model.add(Dense(16,input_dim=7,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(1))
```

```python
model
```

```python
model.compile(optimizer='adam',loss="mean_squared_error",metrics=['mse'])
```

```python
model.fit(X_train,y_train,batch_size=20,epochs=50,validation_split=0.2)
```

```python
model.summary()
```

```python
model2=Sequential()
model2.add(Dense(16,input_dim=7,activation='relu'))
model2.add(Dense(32,activation='relu'))
model2.add(Dense(64,activation='relu'))
model2.add(Dense(128,activation='relu'))
model2.add(Dense(1))
```

```python
import keras
callback = keras.callbacks.EarlyStopping(monitor='loss',patience=5,restore_best_weights=True)
```

```python
model2.compile(optimizer='adam',loss="mean_squared_error",metrics=['mse'])
model2.fit(X_train,y_train,batch_size=15,epochs=25,validation_split=0.25,callbacks=[callback])
```
